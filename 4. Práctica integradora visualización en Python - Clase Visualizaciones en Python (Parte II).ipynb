```python
# Setear para que los gráficos se muestren dentro del mismo notebook:
%matplotlib inline

# Importar librerías:
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 1. Cargar este archivo en Python. Realizar estadísticas descriptivas básicas

```python
salarios= pd.read_csv("Base_desafio_salarios.csv")
#salarios
#salarios.info()
salarios.describe()
```

## 2. Realizar un histograma con los salarios. Qué rango(s) de salarios son los más populares?

```python
salary = salarios.loc[:, 'Salary'] 
#salary
```

```python
x = salary   
fig, ax = plt.subplots()
# Crea la data: distribución normal estandar (0, 1).
ax.hist(x);

# Personalizamos los ejes:
ax.set_xlabel("Salarios")
ax.set_ylabel("N° Empleados");

#Tamaño y resolución de la figura:
fig.set_size_inches(10,7)
fig.set_dpi(100);

# Título de la figura:
fig.suptitle("Histograma de Salarios", fontsize=15, fontweight="bold");

# Agregar líneas:
ax.grid();
```
