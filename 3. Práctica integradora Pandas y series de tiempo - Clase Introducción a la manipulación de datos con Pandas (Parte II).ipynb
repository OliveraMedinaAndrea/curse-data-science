## 1. ¿Cuántos clientes nacidos en Chile (CHI), viven actualmente en Paraguay (PAR)?

### Tratamiento de datos para consigna 1.

```python
# Variable: Pais de nac
    
#Valores posibles y cantidad de registros por c/u:
data_personas.groupby(["Pais Nac"], dropna=False)["ID persona"].count()

# Los valores perdidos se reemplazan por "S/I"
data_personas["Pais Nac"].fillna("S/I", inplace=True)

# Variable: Pais de domicilio (idem anterior)
data_personas.groupby(["Pais Domicilio"], dropna=False)["ID persona"].count()
data_personas["Pais Domicilio"].fillna("S/I", inplace=True)

#Verificación
#data_personas.groupby(["Pais Nac"], dropna=False)["ID persona"].count()
#data_personas.groupby(["Pais Domicilio"], dropna=False)["ID persona"].count()
```

### Código

```python
#Condiciones que aplicamos en la tabla
cond_paisnac_chil= data_personas["Pais Nac"] == "Chi"
cond_paisdomicilio_par = data_personas["Pais Domicilio"] == "Par"

# Aplicamos la condición y filtramos, creando una nueva tabla, para luego contar.
nacidos_en_chile = data_personas.loc[cond_paisnac_chil]

# Condiciones simultáneas
clientes_chil_en_par = data_personas.loc[cond_paisnac_chil & cond_paisdomicilio_par]
```

### Respuesta

```python
print("De los", len(nacidos_en_chile),"clientes nacidos en Chile,", len(clientes_chil_en_par),"viven actualmente en Paraguay.")
```

## 2. ¿Quiénes son los 15 clientes que presentan menor cantidad de compras?


### Tratamiento de datos para consigna 2.

```python
#Valores posibles y cantidad de registros por c/u:
data_personas.groupby(["Nombres"], dropna=False)["ID persona"].count()
data_personas.groupby(["Apellido Mat"], dropna=False)["ID persona"].count()
data_personas.groupby(["Apellido Pat"], dropna=False)["ID persona"].count()

# Los valores perdidos se reemplazan por "S/I" (Sin informacion)
data_personas["Nombres"].fillna("S/I", inplace=True)
data_personas["Apellido Mat"].fillna("S/I", inplace=True)
data_personas["Apellido Pat"].fillna("S/I", inplace=True)
```

### Código

```python
# Creamos la variable Nombre Completo en el Data Frame
data_personas["Nombre Completo"] = data_personas["Nombres"] + " " + data_personas["Apellido Pat"]+ " "+data_personas["Apellido Mat"]

# Creamos una variable con el top 15 de clientes con menor nro. compras
clientes_menor_num_compras = data_personas.sort_values(by = "# de compras", ascending=True, ignore_index=True).iloc[0:15]
```

### Respuesta

```python
# Top 15 de clientes con menor nro. compras
clientes_menor_num_compras.loc[:,["ID persona","# de compras","Nombre Completo"]]
```

## 3. ¿Quiénes son los 50 clientes que presentan menor monto de compras?

### Tratamiento de datos para consigna 3.

```python
# Convertimos este campo (de string) a integer 
data_personas["Monto total en compra"] = data_personas["Monto total en compra"].str.replace('[\$\,\.]', '').astype(int)
```

### Código

```python
# Creamos una variable con el top 50 de clientes con menor monto compras en orden ascendente
clientes_menor_monto_compra=data_personas.sort_values(by = "Monto total en compra", ascending=True, ignore_index=True).iloc[0:50]
```

### Respuesta

```python
# Top 50 de clientes con menor monto compras
clientes_menor_monto_compra.loc[:,["ID persona","Monto total en compra","Nombre Completo"]]
```

## 4. ¿Del total de clientes, qué porcentaje son hombres casados?

### Tratamiento de datos para consigna 4.

```python
# Valores posibles y cantidad de registros por c/u:
data_personas.groupby(["Sexo"], dropna=False)["ID persona"].count()
data_personas.groupby(["Estado civil"], dropna=False)["ID persona"].count()

# Los valores perdidos se reemplazan por "S/I"
data_personas["Sexo"].fillna("S/I", inplace=True)
data_personas["Estado civil"].fillna("S/I", inplace=True)
```

### Código

```python
#Condiciones que aplicamos en la tabla
cond_masculino=data_personas["Sexo"]=="M"
cond_casado = data_personas["Estado civil"]=="Casado/a"

#Condición simultánea
clientes_masculino_casado = data_personas.loc[cond_casado & cond_masculino]
total_porcentual = (len(clientes_masculino_casado)/len(data_personas))*100
```

### Respuesta

```python
print( "Del total de clientes, el ", total_porcentual,"% son hombres casados.")
```

## 5. ¿De los clientes de que viven en Argentina, quiénes son los 20 de mayor Ticket Promedio?

### Código

```python
# Creamos la variable Ticket Promedio en el Data Frame
data_personas["Ticket Promedio"] = data_personas["Monto total en compra"] / data_personas["# de compras"]

# Creamos un nuevo DataFrame de los clientes con domicilio en Arg.
cond_paisdomicilio_arg = data_personas.loc[:,"Pais Domicilio"]=="Arg"
clientes_domicilio_arg = data_personas.loc[cond_paisdomicilio_arg]

# Creamos una variable con el top 20 de clientes con mayor ticket promedio
clientes_mayor_ticketprom= clientes_domicilio_arg.sort_values(by = "Ticket Promedio", ascending=False, ignore_index=True).iloc[0:20]
```

### Respuesta

```python
# Top 20 de clientes con mayor Ticket Promedio:
clientes_mayor_ticketprom.loc[:, ["ID persona","Ticket Promedio","Nombre Completo"]]
```
